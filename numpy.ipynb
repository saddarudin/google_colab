## NumPy (Numerical Python) is the foundational package for numerical computations in Python.
## It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

```python
import numpy as np
```

### Creating arrays

```python
arr = np.array([1,2,3])
arr
```

```python
arr = np.array([[1,2,3],[3,4,5]])
arr
```

```python
arr = np.zeros((3,2))
arr
```

```python
np.ones((3,3))
```

```python
np.random.random((2,2))
```

```python
np.arange(1,10)
```

### linspace(a,b,c) it will generage c numbers in array from a to b which have same difference

```python
np.linspace(2,10,3)
```

## Array shape and reshaping

```python
arr = np.array([[1,2,3],[2,3,4]])
arr.shape
```

```python
arr.reshape(3,2)
```

## Basic Array Operations

```python
arr1 = np.arange(1,5)
arr2 = np.arange(6,10)
arr3 = arr1+arr2
arr3
```

```python
arr1*arr2
```

```python
arr1**2
```

```python
np.sum(arr1)
```

```python
np.mean(arr1)
```

```python
np.min(arr2)
```

```python
np.max(arr1)
```
