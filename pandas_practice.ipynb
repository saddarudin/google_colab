**Calculate the correlation between selling price and kilometers driven.**

**Calculate the average selling price for each fuel type in the dataset.**

```python
fuel_types = df.groupby('fuel')
fuel_types.selling_price.mean()
```

**Find the maximum and minimum selling prices among cars with automatic transmission.**

```python
automatic_transmission=df[df.transmission=='Automatic']
min_price=min(automatic_transmission.selling_price)
max_price=max(automatic_transmission.selling_price)
print(f'Minimum Price is {min_price} & The Maximum price is {max_price}')
# Here we are using isin() because the values can be more than one with minimum
# price if we are sure that output is only one then we use indexing
print(f'\n\nMinimum:\n{automatic_transmission[automatic_transmission.selling_price.isin([min_price])]}')
print(f'\n\nMaximum:\n{automatic_transmission[automatic_transmission.selling_price.isin([max_price])]}')
```

**Group the data by the combination of fuel type and transmission, and find the average selling price for each group.**

```python
grouped_data=df.groupby(['fuel', 'transmission'])
grouped_data.selling_price.mean()
```

**Calculate the total selling price for each year in the dataset.**

```python
yearly=df.groupby('year')
yearly.selling_price.sum()
```
