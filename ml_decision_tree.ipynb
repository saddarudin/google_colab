```python
df1 = pd.read_csv('titanic.csv')
df1.head(10)
```

```python
x = df1[['Pclass','Sex','Age','Fare']]
x
```

```python
y = df1.Survived
y
```

```python
y.value_counts()
```

```python
y.isna().sum()
```

```python
x.isna().sum()
```

```python
x.info()
```

```python
x.fillna({
    'Age':x.Age.mean()
},inplace=True)
x.isna().sum()
```

```python
x.Sex.value_counts()
```

```python
x.replace({
    'Sex':{
        'male':1,
        'female':0
    }
},inplace=True)
x.info()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train.shape,x_test.shape
```

```python
x.head()
```

```python
model1 = DecisionTreeClassifier()
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model1.predict([[3,1,36,50]])
```

```python
y_pred = model1.predict(x_test)
y_pred[0:5]
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```
