### Array Range Function
### arange(a,b,c) where a=start, b=end,c=steps

```python
arr = np.arange(1,10)
arr
```

```python
# odd numbers less than 20
arr = np.arange(1,20,2)
arr
```

```python
print('Even numbers less than 21')
arr = np.arange(0,21,2)
arr
```

### Reshape Function

```python
original_array = np.array([[1,2,3],[3,4,5]])
print(original_array)
print(original_array.shape)
```

```python
reshaped_array = original_array.reshape(3,2)
print(reshaped_array)
print(reshaped_array.shape)
```

### Flattening Array using ravel function

### Ravel will not alter original array

```python
flatten_array = reshaped_array.ravel()
```

```python
flatten_array
```

```python
reshaped_array
```

```python
min(flatten_array)
```

```python
flatten_array.min()
```
