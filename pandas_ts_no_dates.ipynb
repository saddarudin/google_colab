```python
import pandas as pd
```

```python
df = pd.read_csv('aapl_no_dates.csv')
df
```

## When there are no dates you can add using date_range function

```python
# here B means business days
rng = pd.date_range(start='01-05-2025',periods=22,freq='B')
rng
```

```python
df.set_index(rng,inplace=True)
df.head()
```

## Creating pandas Series with random number and range of dates

```python
import numpy as np
ts = pd.Series(np.random.randint(1,30,len(rng)),index=rng,)
ts
```
