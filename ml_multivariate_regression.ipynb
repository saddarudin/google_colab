```python
import numpy as np
import pandas as pd
import sklearn.linear_model as lm
```

```python
df = pd.read_csv('home_prices.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
import math
df.fillna({
    'bedrooms': math.floor(df.bedrooms.mean())
},inplace=True)
df
```

```python
model = lm.LinearRegression()
model.fit(df[['area','bedrooms','age']],df.price)
```

```python
model.predict([[4100,7,10]])
```

```python
from word2number import w2n
for i in range(len(df1.experience)):
    df1.experience[i] = w2n.word_to_num(df1.experience[i])

df1
```

```python
model1 = lm.LinearRegression()
model1.fit(df1[['experience','test_score(out of 10)','interview_score(out of 10)']],df1['salary($)'])
```
