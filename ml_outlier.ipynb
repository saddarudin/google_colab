## Removing outliers using percentile. Method is quantile

```python
import pandas as pd
```

```python
df = pd.read_csv('heights.csv')
df.head()
```

```python
max_threshold=df.height.quantile(0.99)
max_threshold
```

```python
df[df.height>max_threshold]
```

```python
min_threshold = df.height.quantile(0.01)
min_threshold
```

```python
df[df.height<min_threshold]
```

```python
df.shape
```

```python
df = df[(df.height>min_threshold) & (df.height<max_threshold)]
df.shape
```
