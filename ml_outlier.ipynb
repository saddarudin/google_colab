## Removing outliers using percentile. Method is quantile

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heights.csv')
df.head()
```

```python
max_threshold=df.height.quantile(0.99)
max_threshold
```

```python
df[df.height>max_threshold]
```

```python
min_threshold = df.height.quantile(0.01)
min_threshold
```

```python
df[df.height<min_threshold]
```

```python
df.shape
```

```python
df = df[(df.height>min_threshold) & (df.height<max_threshold)]
df.shape
```

## Removing Outlier using Standard Deviation

```python
df2 = pd.read_csv('heights.csv')
df2.head()
```

```python
plt.hist(df2.height,bins=20,rwidth=0.8)
plt.xlabel('Height(inches)')
plt.ylabel('Frequency')
plt.show()
```

```python
upper_limit = df2.height.mean()+3*df2.height.std()
upper_limit
```

```python
lower_limit = df2.height.mean()-3*df2.height.std()
lower_limit
```

```python
df2[(df2.height<lower_limit) | (df2.height>upper_limit)]
```

```python
df2.shape
```

```python
df2 = df2[(df2.height>lower_limit) & (df2.height<upper_limit)]
df2.shape
```
