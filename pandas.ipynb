## Series: One-dimensional labeled array, similar to a list or dictionary.

## Data Cleaning

### checking for missing values

```python
df.isnull()
```

### dropping null values

```python
df.dropna()
```

### Filling missing values

```python
df.fillna('something')
```

### removing duplicates

```python
df.drop_duplicates()
```

### replacing values

```python
df.replace({1000:1300},inplace=True)
```

```python
df.head()
```

## Renaming columns

```python
df.rename(columns={'Unnamed: 0':'S#',},inplace=True)
df.head()
```

### Adding new column

```python
df['price/area'] = df['price']/df['area']
df.head()
```

### dropping columns

```python
df.drop('S#',axis='columns',inplace=True)
df.head()
```

### Filterring row

```python
df[df.area>1200]
```

### Sorting values

```python
df.sort_values('area',ascending=True,inplace=True)
df.head()
```

###Group by

```python
grouped_data=df.groupby('price')
grouped_data.describe()
```

## Merging data

```python
left = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob']})
right = pd.DataFrame({'ID': [1, 2], 'City': ['New York', 'Paris']})

# Merge on common column 'ID'
merged_df = pd.merge(left, right, on='ID')
print(merged_df)
```
