```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from transformers import pipeline
```

```python
senti_model = pipeline(task='sentiment-analysis',model="cardiffnlp/twitter-roberta-base-sentiment-latest",device='cuda')
```

```python
senti_model('I love you')[0]['label']
```

```python
df = pd.read_csv('reviews_data.csv')
df.head()
```

```python
reviews = pd.DataFrame(df.Review)
reviews.head()
```

```python
reviews.Review
```

```python
reviews['prediction'] = reviews.Review.apply(lambda x: senti_model(str(x))[0]['label'])
reviews.head()
```

```python
reviews.prediction.unique()
```

```python
reviews.prediction.value_counts()
```

```python
positive = reviews[reviews.prediction == 'positive']
negative = reviews[reviews.prediction == 'negative']
neutral = reviews[reviews.prediction == 'neutral']
positive.head()
```
