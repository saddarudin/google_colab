```python
import pandas as pd
```

```python
df = pd.read_csv('aapl.csv',parse_dates=['Date'],index_col='Date')
df.head()
```

```python
df.index
```

```python
df = df.sort_index()
df.head()
```

## Retrieve January 2017 Stock Prices

```python
df.loc["2017-01"]
```

## Average price of apple stock in january 2017

```python
df.loc['2017-01'].Close.mean()
```

## Price in range of dates

```python
df.loc["2017-01-01":"2017-01-15"]
```

## Monthly average close balance

```python
result = df.Close.resample('ME').mean()
result
```

```python
import matplotlib.pyplot as plt
plt.plot(result)
plt.show()
```
