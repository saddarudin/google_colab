```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
digits = load_digits()
dir(digits)
```

```python
digits.feature_names[0]
```

```python
digits.data[0]
```

```python
plt.matshow(digits.images[0])
```

```python
digits.target[0:5]
```

```python
digits.target_names[0:5]
```

```python
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.matshow(digits.images[1736])
```

```python
model.predict([digits.data[1736]])
```

```python
plt.matshow(digits.images[338])
```

```python
model.predict([digits.data[338]])
```

```python
for i in range(5):
  plt.matshow(digits.images[334+i])
```

```python
digits.target[334:339]
```

```python
pred = digits.data[334:339]
```

```python
model.predict(pred)
```

### Generating Confusino Metrix

```python
y_pred = model.predict(x_test)
y_pred[0:5]
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
