```python
import pandas as pd
```

```python
df_pk = pd.DataFrame({
    'city':['Dadu','Jamshoro','Larkana','Karachi','Sukkur'],
    'temperature':[40,35,41,31,39],
    'humidity':[80,60,70,50,65]
})
df_us = pd.DataFrame({
    'city':['Chicago','New York','San Fransico','Times Square','London Tower'],
    'temperature':[21,14,35,32,33],
    'humidity':[68,65,75,70,75]
})
```

```python
df1 = pd.DataFrame({
    'city':['Dadu','Jamshoro','Larkana'],
    'temperature':[51,41,50]
})
df1
```

### For example you have some different columns

```python
df1 = pd.DataFrame({
    'city':['Dadu','Jamshoro','Karachi','Hyderabad'],
    'temperature':[51,41,50,30]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['Jamshoro','Dadu','Larkana','Sukkur'],
    'humidity':[31,35,41,50]
})
df2
```

## This is by defaule inner join

```python
pd.merge(df1,df2,on='city')
```

## This is outer join (it is similar to union)

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

## When repeated columns are present

```python
df1 = pd.DataFrame({
    'city':['Dadu','Jamshoro','Karachi','Hyderabad'],
    'temperature':[51,41,50,30],
    'humidity':[50,60,70,80]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['Jamshoro','Dadu','Larkana','Sukkur'],
    'temperature':[30,40,50,20],
    'humidity':[30,40,50,60]
})
df2
```

```python
pd.merge(df1,df2,on='city')
```

## You notice that it automatically assigns x and y suffixes

## You can apply your own suffixes

```python
pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
```
